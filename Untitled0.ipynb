```python
def adicionar_tarefa(nome_tarefa, descricao="Adicionado pelo Agente"):
    # 1. Busca o quadro chamado 'DIO'
    boards = client.list_boards()
    board_dio = next((b for b in boards if b.name == 'DIO'), None)

    if not board_dio:
        return "Quadro 'DIO' não encontrado. Crie um quadro com esse nome no Trello!"

    # 2. Busca a lista 'A Fazer' (ou a primeira lista que encontrar)
    listas = board_dio.list_lists()
    lista_alvo = next((l for l in listas if "fazer" in l.name.lower()), listas[0])

    # 3. Adiciona o cartão
    novo_card = lista_alvo.add_card(nome_tarefa, desc=descricao)
    return f"Tarefa '{nome_tarefa}' adicionada com sucesso no quadro DIO!"

# Testando a habilidade manualmente
print(adicionar_tarefa("Comprar ração para o dog"))
```

```python
def listar_todas_as_tarefas():
    # 1. Busca o quadro DIO novamente para garantir a conexão
    boards = client.list_boards()
    board_dio = next((b for b in boards if b.name == 'DIO'), None)

    if not board_dio:
        return "Quadro 'DIO' não encontrado. Verifique o nome no Trello!"

    resumo = "📋 --- STATUS ATUAL DO QUADRO DIO --- 📋\n"

    # 2. Percorre cada lista (coluna) do quadro
    for lista in board_dio.list_lists():
        resumo += f"\n🔹 {lista.name.upper()}:"
        cards = lista.list_cards()

        if not cards:
            resumo += "\n  (Vazia)\n"
        else:
            for card in cards:
                resumo += f"\n  ✅ {card.name}"
            resumo += "\n"

    return resumo

# Executando a função para ver o resultado
print(listar_todas_as_tarefas())
```

```python
def mover_tarefa(nome_tarefa):
    boards = client.list_boards()
    board_dio = next((b for b in boards if b.name == 'DIO'), None)

    if not board_dio:
        return "Quadro 'DIO' não encontrado."

    # Mapeamento dos status (a ordem importa!)
    ordem_status = ["A Fazer", "Em Andamento", "Concluído"]

    # 1. Encontrar o cartão em qualquer lista
    card_alvo = None
    lista_atual = None

    for lista in board_dio.list_lists():
        cards = lista.list_cards()
        card_alvo = next((c for c in cards if c.name.lower() == nome_tarefa.lower()), None)
        if card_alvo:
            lista_atual = lista
            break

    if not card_alvo:
        return f"Tarefa '{nome_tarefa}' não encontrada para mover."

    # 2. Descobrir para qual lista mover
    try:
        indice_atual = -1
        for i, nome in enumerate(ordem_status):
            if nome.lower() in lista_atual.name.lower():
                indice_atual = i
                break

        if indice_atual == -1 or indice_atual == len(ordem_status) - 1:
            return f"A tarefa já está no status final ou status atual é inválido."

        proximo_status = ordem_status[indice_atual + 1]

        # 3. Localizar a lista de destino
        lista_destino = next((l for l in board_dio.list_lists() if proximo_status.lower() in l.name.lower()), None)

        if lista_destino:
            card_alvo.change_list(lista_destino.id)
            return f"Sucesso! '{nome_tarefa}' movida de '{lista_atual.name}' para '{lista_destino.name}'."
        else:
            return "Lista de destino não encontrada."

    except Exception as e:
        return f"Erro ao mover: {e}"

# Teste o movimento!
print(mover_tarefa("Comprar ração para o dog"))
```

```python
def processar_comando_agente(texto_usuario):
    # O "System Prompt" ensina o Gemini como ele deve se comportar
    prompt_sistema = f"""
    Você é um Agente Organizador de Tarefas. Você tem acesso a três funções:
    1. adicionar_tarefa(nome_tarefa): Use quando o usuário quiser criar algo novo.
    2. listar_todas_as_tarefas(): Use quando o usuário quiser saber o que tem para fazer.
    3. mover_tarefa(nome_tarefa): Use quando o usuário disser que terminou algo ou quer avançar o status.

    Texto do usuário: "{texto_usuario}"

    Responda APENAS com o nome da função e os parâmetros necessários em formato Python,
    exemplo: adicionar_tarefa("Lavar o carro") ou listar_todas_as_tarefas().
    Se não entender, peça desculpas.
    """

    # O Gemini decide o que fazer
    resposta_ia = model.generate_content(prompt_sistema).text.strip()

    try:
        # Aqui o Python executa a decisão da IA
        resultado = eval(resposta_ia)
        return f"🤖 Agente diz: {resultado}"
    except Exception as e:
        return f"🤖 Agente: Não entendi bem, pode repetir? (Erro: {resposta_ia})"

# --- LOOP DE CONVERSA ---
print("Agente DIO Online! (Digite 'sair' para encerrar)")
while True:
    user_input = input("Você: ")
    if user_input.lower() == 'sair':
        break
    print(processar_comando_agente(user_input))
```
